```python
import cv2

import numpy as np

from tensorflow.keras.models import load_model

 

# Load the saved model

model = load_model('drowsiness_detection_model_rnn.h5')

 

def preprocess_image(image_path, target_size=(227, 227)):

    # Load the image

    image = cv2.imread(image_path)

    # Resize the image to the target size

    image = cv2.resize(image, target_size)

    # Normalize the pixel values to be between 0 and 1

    image = image / 255.0

    # Expand dimensions to match the input shape of the model

    image = np.expand_dims(image, axis=0)

    return image

 

def predict_drowsiness(image_path):

    # Preprocess the image

    processed_image = preprocess_image(image_path)

    # Make predictions

    predictions = model.predict(processed_image)

    # Convert the prediction to a human-readable label

    if predictions[0][0] > 0.5:

        return "Drowsy"

    else:

        return "Not Drowsy"

 

# Example usage

new_image_path = './test/A1329.png'

prediction = predict_drowsiness(new_image_path)

print(f'The prediction for the image is: {prediction}')
```
